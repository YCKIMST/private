```python
m = folium.Map(
    location = [33.36309860282704, 126.52950979710285],
    #tiles = 'Stamen Terrain',
    zoom_start=10
)

tooltip = 'Click!!'

folium.Marker(
    [33.36309860282704, 126.52950979710285],
    popup = '<strong>한란산</strong>',
    tooltip = tooltip,
    icon = folium.Icon(color='#ffffff', icon = 'info-sign')  #ff(R)ff(G)ff(B)
).add_to(m)

folium.Marker(
    [33.51064606468837, 126.49139631334086],
    popup = '<strong>제주국제공항</strong>',
    tooltip = tooltip,
    icon = folium.Icon(color='green', icon = 'bookmark')  #ff(R)ff(G)ff(B)
).add_to(m)

folium.Marker(
    [33.48784010820233, 126.48199828273358],
    popup = '<strong>그랜드 하얏트 제주</strong>',
    tooltip = tooltip
).add_to(m)




m
```
